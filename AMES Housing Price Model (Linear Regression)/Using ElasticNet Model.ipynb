The data has been pre-processed and missing values have been treated

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data

```python
df = pd.read_csv("../DATA/AMES_Final_DF.csv")
```

## Exploring Dataset

```python
df.head()
```

```python
df['SalePrice'].describe()
```

**TASK: The label we are trying to predict is the SalePrice column. Separate out the data into X features and y labels**

```python
X = df.drop('SalePrice',axis=1)
y = df['SalePrice']
y
```

**Training Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.1,random_state=101)
```

**The dataset features has a variety of scales and units. For optimal regression performance, we scale the X features.**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_test = scaler.transform(X_test)
scaled_X_train = scaler.transform(X_train)
```

**We will use an Elastic Net model to apply both Lasso and Ridge Regressions**

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet(max_iter=1000000)
```

**The Elastic Net model has two main parameters, alpha and the L1 ratio.**

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[0.3,0.5,0.8,0.95,0.99,1]}
```

**Finding optimal hyperparameters using GridSearchCV - 10 fold Cross-Validation**

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_model,param_grid=param_grid
                          ,cv=10,verbose=2,scoring='neg_mean_squared_error')
```

```python
grid_model.fit(scaled_X_train,y_train)
```

**Best Combination of Parameters**

```python
res2 = grid_model.cv_results_
res_def_iter2 = pd.DataFrame(res2).sort_values(by='mean_test_score')[['param_alpha','param_l1_ratio','mean_test_score']]
```

```python
res_def_iter2
```

```python
coef = pd.Series(grid_model.best_estimator_.coef_)
```

```python
coef = coef[coef == 0].index
coef
```

```python
## Number of coefficients which are dropped
len(coef)
```

```python
## Names of Parameters which became 0
drop_columns = df.columns[coef]
```

```python
drop_columns
```

```python
df_with_important_features = df.drop(drop_columns,axis=1  )
```

```python
df_with_important_features
```

**Evaluating model's performance on the unseen 10% scaled test set.**

```python
test_predictions = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
MAE
```

```python
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)
RMSE
```
